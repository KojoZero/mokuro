Before running this, inside Google Drive, create a folder named "mokuro-ocr" and create a folder for each individual manga **series** that will be ocr'ed. Then place the individual manga **volumes** inside its respective manga series folder. The directory should look like below:

```
MyDrive/
├─mokuro-ocr/
  ├─manga/
    ├─vol1/
```
or

```
MyDrive/
├─mokuro-ocr/
  ├─manga/
    ├─vol1.zip
```

For the commands below, replace ```<manga>``` with your manga series, and ```<vol1>``` with the individual volume you want to OCR.

To run on individual volume folders:

````python
!mokuro "/content/drive/MyDrive/mokuro-ocr/<manga>/<vol1>"
````

To run on individual volume zip files:

````python
!mokuro "/content/drive/MyDrive/mokuro-ocr/<manga>/<vol1>.zip"
````

Run on a directory containing multiple volumes (either zip or folder format):

````python
!mokuro --parent_dir "/content/drive/MyDrive/mokuro-ocr/<manga>"
````
